```python
import holidays
import pandas as pd
```

```python

# List to hold holiday data
holiday_data = []


# Example: Pull US holidays for 2021-2025
for year in range(2023, 2026):
    us_holidays = holidays.US(years=year)
    for date, name in sorted(us_holidays.items()):
        holiday_data.append({"date": date, "holiday_name": name})


```

```python
df_holidays = pd.DataFrame(holiday_data)
```
